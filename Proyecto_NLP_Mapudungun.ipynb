```python
def file_to_translation_list(path: str) -> list:
  ''' Limpiamos el file y generamos lista que contiene
      frases en Mapuzungun y Espanol '''
  with open(path, 'r', encoding='utf-8') as file:
    lines = file.readlines()
    M_E_list = []
    for line in lines:
      if line[0] not in ("M", "C"):
        continue
      else:
        M_E_list.append(line.rstrip())
  return M_E_list
```

```python
def translation_list_to_tuples(phrases: list) -> list:
  ''' Generamos lista de tuplas (Mapuzungun, Castellano)
      Función generada con chatgpt '''

  cleaned_phrases = []

  for phrase in phrases:
    if phrase.startswith("M:"):
      cleaned_phrases.append((phrase[3:].strip(), None))
    elif phrase.startswith("C:"):
      if cleaned_phrases and cleaned_phrases[-1][1] is None:
        cleaned_phrases[-1] = (cleaned_phrases[-1][0], phrase[3:].strip())
      else:
        cleaned_phrases.append((None, phrase[3:].strip()))

  return [phrase for phrase in cleaned_phrases if phrase[0] is not None and phrase[1] is not None]
```

```python
def tuples_to_dataframe(paired_phrases: list) -> pd.DataFrame:
    ''' Transforma lista de tuplas en dataframe '''
    df = pd.DataFrame(paired_phrases, columns=["Mapudungun", "Español"])
    return df
```

```python
def join_data_frames(path: str):
    ''' Unimos los dataframes de los diferentes archivos '''
    filenames = next(walk(path), (None, None, []))[2]

    with open('dataset_splits/mt/training_files.txt', 'r', encoding='utf-8') as archive:
        training_filenames = [line.strip() + '.txt' for line in archive.readlines()]

    with open('dataset_splits/mt/test_files.txt', 'r', encoding='utf-8') as archive:
        test_filenames = [line.strip() + '.txt' for line in archive.readlines()]

    with open('dataset_splits/mt/dev_files.txt', 'r', encoding='utf-8') as archive:
        val_filenames = [line.strip() + '.txt' for line in archive.readlines()]

    training_frames, test_frames, val_frames = [], [], []

    for archiv in filenames:
        try:
            lst = file_to_translation_list(os.path.join(path, archiv))
            print("Processing:", os.path.join(path, archiv))
            tuples_list = translation_list_to_tuples(lst)
            df = tuples_to_dataframe(tuples_list)
            if archiv in training_filenames:
                training_frames.append(df)
            if archiv in test_filenames:
                test_frames.append(df)
            if archiv in val_filenames:
                val_frames.append(df)
        except Exception as e:
            print(f"Error processing {archiv}: {e}")

    print(len(training_frames), len(test_frames), len(val_frames))

    return pd.concat(training_frames, ignore_index=True), pd.concat(test_frames, ignore_index=True), pd.concat(val_frames, ignore_index=True)
```

```python
dataset_directory = "translation-clean"
train_df, test_df, val_df = join_data_frames(dataset_directory)
```

```python
train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)
test_dataset = Dataset.from_pandas(test_df)
```

```python
dataset_dict = DatasetDict({
    'train': train_dataset,
    'validation': val_dataset,
    'test': test_dataset
})
```

```python
model_checkpoint = "Helsinki-NLP/opus-mt-es-en"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

def preprocess_function(examples):
    inputs = examples["Mapudungun"]
    targets = examples["Español"]
    model_inputs = tokenizer(inputs, max_length=128, truncation=True, padding="max_length")
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True, padding="max_length")
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_datasets = dataset_dict.map(preprocess_function, batched=True)
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
)
```

```python
trainer.train()
```

```python
model_save_path = "./trained_model"
tokenizer_save_path = "./trained_tokenizer"

model.save_pretrained(model_save_path)

tokenizer.save_pretrained(tokenizer_save_path)
```

```python
def evaluate_translations_rouge(test_df, token_length, num_samples=100):
    # Filter examples by token length
    filtered_df = test_df[test_df['Mapudungun'].apply(lambda x: len(tokenizer(x)['input_ids']) == token_length)].sample(n=num_samples)

    # Translate the filtered examples
    translations = filtered_df['Mapudungun'].apply(translate)

    # Calculate ROUGE score
    metric = load_metric("rouge")
    references = filtered_df['Español'].tolist()
    predictions = translations.tolist()
    rouge_score = metric.compute(predictions=predictions, references=references)

    return rouge_score
```

```python
def evaluate_translations_bleu_length_less_than_n(test_df, n, num_samples=100):
    # Filter examples with less than n tokens
    filtered_df = test_df[test_df["n_words"] < n].head(num_samples)

    # Translate the filtered examples
    translations = filtered_df['Mapudungun'].apply(translate)

    # Calculate BLEU score
    metric = load_metric("bleu")
    references = filtered_df['Español'].apply(lambda x: [x.split()]).tolist()  # BLEU expects list of references
    predictions = translations.apply(lambda x: x.split()).tolist()  # BLEU expects tokenized predictions
    bleu_scores = metric.compute(predictions=predictions, references=references)

    # Calculate average BLEU score
    avg_bleu_score = bleu_scores['bleu']

    return avg_bleu_score
```

```python
def evaluate_translations_bleu_length_more_than_n(test_df, n, num_samples=100):
    # Filter examples with length more than or equal to n tokens
    filtered_df = test_df[test_df["n_words"] >= n].head(num_samples)

    # Translate the filtered examples
    translations = filtered_df['Mapudungun'].apply(translate)

    # Calculate BLEU score
    metric = load_metric("bleu")
    references = filtered_df['Español'].apply(lambda x: [x.split()]).tolist()  # BLEU expects list of references
    predictions = translations.apply(lambda x: x.split()).tolist()  # BLEU expects tokenized predictions
    bleu_scores = metric.compute(predictions=predictions, references=references)

    # Calculate average BLEU score
    avg_bleu_score = bleu_scores['bleu']

    return avg_bleu_score
```

```python
def evaluate_translations_bleu_length_between_m_and_n(test_df, m, n, num_samples=100):
    # Filter examples with length between m and n tokens
    filtered_df = test_df[(test_df["n_words"] >= m) & (test_df["n_words"] < n)].head(num_samples)

    # Translate the filtered examples
    translations = filtered_df['Mapudungun'].apply(translate)

    # Calculate BLEU score
    metric = load_metric("bleu")
    references = filtered_df['Español'].apply(lambda x: [x.split()]).tolist()  # BLEU expects list of references
    predictions = translations.apply(lambda x: x.split()).tolist()  # BLEU expects tokenized predictions
    bleu_scores = metric.compute(predictions=predictions, references=references)

    # Calculate average BLEU score
    avg_bleu_score = bleu_scores['bleu']

    return avg_bleu_score
```

```python
def calculate_rouge(row):
    prediction = translate(row['Mapudungun'])

    scores = scorer.score(row['Español'], prediction)

    return pd.Series({
        'rouge1_f1': scores['rouge1'].fmeasure,
        'rouge2_f1': scores['rouge2'].fmeasure,
        'rougeL_f1': scores['rougeL'].fmeasure
    })
```
